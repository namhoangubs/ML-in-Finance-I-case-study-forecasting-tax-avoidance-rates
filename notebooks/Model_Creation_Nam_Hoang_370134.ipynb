```python
rfr = rfg(n_jobs=-1, max_depth=5)
rfecv_rf = RFECV(rfr, step=1, cv=3)
rfecv_rf.fit(df.loc[:, x_col], df.loc[:, y_col])
```

```python
X_rf = features_rf[mask_rf].tolist()
with open('C:/Users/Lenovo/Desktop/Machine Learning 1/project 2/random_forest_variables.pkl', 'wb') as file:
    pickle.dump(X_rf, file)
```

```python
def objective_rf(trial, X, y):
    params = {'n_estimators': trial.suggest_int('n_estimators', 50, 1000, log=True), 
              'max_depth': trial.suggest_int('max_depth', 2, 100, log=True),
              'min_samples_split': trial.suggest_int('min_samples_split', 2, 20, log=True),
              'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10, log=True),              
             }
    model = make_pipeline(rfg(**params, random_state = 16))
    scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
    return np.mean(scores)

study_rf = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=0))
study_rf.optimize(functools.partial(objective_rf, X=df[X_rf], y=df[y_col]), n_trials=30, n_jobs=8)
```

```python
def objective_ecv(trial, X, y):
    params = {
        'l1_ratio': trial.suggest_float('l1_ratio', 0.0, 1.0),
        'alphas': [trial.suggest_float(f'alpha_{i}', 1e-5, 1e-1) for i in range(5)],  # Specify multiple alpha values
        'cv': 5,  # Number of cross-validation folds
        }
    model = make_pipeline(ElasticNetCV(**params, random_state = 16))
    scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
    return np.mean(scores)

study_ecv = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=0))
study_ecv.optimize(functools.partial(objective_ecv, X=df[X_ecv], y=df[y_col]), n_trials=100, n_jobs=8)
```
